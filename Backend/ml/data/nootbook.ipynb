```python
import xgboost as xgb
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
def train_and_eval(X_train, X_test, y_train, y_test):
    model = xgb.XGBClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=4,
        random_state=42,
        eval_metric='logloss'
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)
```

```python
X_train = X_train.copy()
X_test  = X_test.copy()

# Total Skills
X_train['Total_Skills'] = (
    0.4*X_train['Coding_Skills'] +
    0.1*X_train['Projects'] +
    0.1*X_train['Internships'] +
    0.4*X_train['CGPA']
)

X_test['Total_Skills'] = (
    0.4*X_test['Coding_Skills'] +
    0.1*X_test['Projects'] +
    0.1*X_test['Internships'] +
    0.4*X_test['CGPA']
)

# Skill Power Balance
X_train['Skill_P_B'] = (
    X_train['Coding_Skills'] +
    X_train['Projects'] +
    X_train['Internships'] -
    X_train['Backlogs']
)

X_test['Skill_P_B'] = (
    X_test['Coding_Skills'] +
    X_test['Projects'] +
    X_test['Internships'] -
    X_test['Backlogs']
)
```

```python
baseline_features = [
    'CGPA','Coding_Skills','Projects','Internships',
    'Backlogs','Communication_Skills','Aptitude_Test_Score'
]

acc_e1 = train_and_eval(
    X_train[baseline_features],
    X_test[baseline_features],
    y_train, y_test
)
```

```python
features_e2 = baseline_features + ['Total_Skills']

acc_e2 = train_and_eval(
    X_train[features_e2],
    X_test[features_e2],
    y_train, y_test
)
```

```python
engineered_features = ['Total_Skills', 'Skill_P_B']

acc_e3 = train_and_eval(
    X_train[engineered_features],
    X_test[engineered_features],
    y_train, y_test
)
```

```python
features_e4 = baseline_features  # explicitly no Total_Skills

acc_e4 = train_and_eval(
    X_train[features_e4],
    X_test[features_e4],
    y_train, y_test
)
```

```python
results = pd.DataFrame({
    'Experiment': [
        'Baseline (All Original)',
        'All + Total_Skills',
        'Only Engineered',
        'Without Total_Skills'
    ],
    'Accuracy': [acc_e1, acc_e2, acc_e3, acc_e4]
})

print(results)
```

```python
def add_academic_risk(X):
    X = X.copy()
    X['Academic_Risk'] = X['Backlogs'] / (X['CGPA'] + 0.1)
    return X
```

```python
def add_practical_exposure(X):
    X = X.copy()
    X['Practical_Exposure'] = (
        X['Projects'] +
        X['Internships'] +
        X['Certifications']
    )
    return X
```

```python
def add_skill_readiness(X):
    X = X.copy()
    X['Skill_Readiness'] = (
        0.6 * X['Coding_Skills'] +
        0.4 * X['Aptitude_Test_Score']
    )
    return X
```

```python
import numpy as np

def add_salary_column(df):
    """
    Injects a Salary column if not present.
    Salary is based on CGPA, skills, and experience.
    Non-placed students get Salary = 0.
    """

    if 'Salary' not in df.columns:

        assert 'CGPA' in df.columns
        assert 'Placement_Status' in df.columns

        base_salary = (
            df['CGPA'] * 60000 +
            df['Coding_Skills'] * 20000 +
            df['Communication_Skills'] * 15000 +
            df['Projects'] * 10000 +
            df['Internships'] * 12000 +
            np.random.normal(0, 15000, size=len(df))
        )

        df['Salary'] = base_salary
        df.loc[df['Placement_Status'] == 0, 'Salary'] = 0
        df['Salary'] = df['Salary'].clip(lower=0)

    return df
```
